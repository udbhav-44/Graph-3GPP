```python
from llama_index.readers.file import DocxReader
# from pathlib import Path
from llama_index.llms.openai import OpenAI
from DataModel.datamodel import DataModel
import json
# import aiofiles
from dotenv import load_dotenv
load_dotenv()

llm = OpenAI(model="gpt-4o-mini")
```

```python
from langchain_community.document_loaders import UnstructuredWordDocumentLoader

loader = UnstructuredWordDocumentLoader("source_docs/2.docx")
data = loader.load()

print(data[0].page_content)
```

```python



# print(data)

sllm = llm.as_structured_llm(DataModel)

response = sllm.complete(data[0].page_content)


```

```python
from DataModel.datamodel import DataModelEncoder
with open("output.json", "w") as f:
    json.dump(response.raw, f, indent=4, cls=DataModelEncoder)
    
```
